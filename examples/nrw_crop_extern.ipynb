```python
nrw_crops_2021 = gpd.read_file("~/Downloads/eurocrop/NRW/DE_NRW_2021_EC21.shp")
nrw_crops_2021
```

```python
nrw_crops_2021["points"]= nrw_crops_2021.centroid
nrw_crops_2021.set_geometry("points", inplace=True)
nrw_crops_2021
```

```python
nrw_crops_2021_points_code = nrw_crops_2021[["ID", "points", "CODE"]]
nrw_crops_2021_points_code
```

```python
group_sample = nrw_crops_2021_points_code.groupby("CODE").sample(10, random_state=1337, replace=True)
group_sample.reset_index(inplace=True)
group_sample
```

```python
bounds_array = group_sample.total_bounds
xmin = bounds_array[0]
ymin = bounds_array[1]
xmax = bounds_array[2]
ymax = bounds_array[3]
```

```python
ge_res = ge.upload_dataframe(group_sample, "group_sample")
ge_res
```

```python
ge_point_source = ge.unstable.workflow_operators.OgrSource(str(ge_res))
ge_point_source.to_dict()
```

```python
import geoengine
    
projected_points = geoengine.unstable.workflow_operators.Reprojection(ge_point_source, target_spatial_reference="EPSG:32632")

s2_cloud_free_operator = geoengine.unstable.workflow_blueprints.sentinel2_cloud_free_band("B02")

# We could also leave out the scaling and use the I64 data directly
s2_cloud_free_operator = geoengine.unstable.workflow_operators.RasterTypeConversion(
    source=s2_cloud_free_operator,
    output_data_type="F32"
)
s2_cloud_free_operator = geoengine.unstable.workflow_operators.RasterScaling(
    source=s2_cloud_free_operator,
    slope = 0.0001,
    offset = -0.1, # this should be -0.1 but the values are too small?
    scaling_mode="mulSlopeAddOffset"
)

monthly_s2_cloud_free_operator = geoengine.unstable.workflow_operators.TemporalRasterAggregation(
    source=s2_cloud_free_operator,
    aggregation_type="mean",
    granularity="days",
    window_size=15,
    ignore_no_data=True,
)
points_with_s2_cloud_free = geoengine.unstable.workflow_operators.RasterVectorJoin(
    raster_sources=[monthly_s2_cloud_free_operator],
    vector_source=projected_points,
    new_column_names=["B02"],
)

points_with_s2_cloud_free.to_dict()
        
```

```python
workflow = geoengine.register_workflow(points_with_s2_cloud_free.to_workflow_dict())
workflow
```

```python
import datetime

start_dt = datetime.datetime(2021, 6, 1, 0, 0, 0, tzinfo=datetime.timezone.utc)
end_dt = datetime.datetime(2021, 6, 14, 0, 0, 0, tzinfo=datetime.timezone.utc)


gp_res = await workflow.vector_stream_into_geopandas(
    geoengine.QueryRectangle(
        spatial_bounds=geoengine.BoundingBox2D(
            xmin=xmin,
            ymin=ymin,
            xmax=xmax,
            ymax=ymax,
        ),
        time_interval=geoengine.TimeInterval(
            start=start_dt,
            end=end_dt,
        ),
        resolution=geoengine.SpatialResolution(
            100.0,
            100.0,
        ),
        srs="EPSG:32632",
))

gp_res.to_file("gp_res.geojson", driver="GeoJSON")
gp_res
```
