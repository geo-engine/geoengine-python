```python
import geoengine
    
def multiple_s2_cloud_free_bands(bands, granularity="days", window_size=1, aggregation_type="months"):
    valid_sentinel_bands = ["B01", "B02", "B03", "B04", "B05", "B06", "B07", "B08", "B8A", "B11", "B12", "SCL"]
    operators = {}

    for band in bands:
        if band not in valid_sentinel_bands and band != "NDVI":
            raise ValueError(f"Invalid band: {band}")
        
        if band == "NDVI":
            s2_cloud_free_operator = geoengine.unstable.workflow_blueprints.sentinel2_cloud_free_ndvi()
        else:
            s2_cloud_free_operator = geoengine.unstable.workflow_blueprints.sentinel2_cloud_free_band(band)

            # We could also leave out the scaling and use the I64 data directly
            #s2_cloud_free_operator = geoengine.unstable.workflow_operators.RasterTypeConversion(
            #    source=s2_cloud_free_operator,
            #    output_data_type="F32"
            #)
            #s2_cloud_free_operator = geoengine.unstable.workflow_operators.RasterScaling(
            #    source=s2_cloud_free_operator,
            #    slope = 0.0001,
            #    offset = -0.1, # this should be -0.1 but the values are too small?
            #    scaling_mode="mulSlopeAddOffset"
            #)

        monthly_s2_cloud_free_operator = geoengine.unstable.workflow_operators.TemporalRasterAggregation(
            source=s2_cloud_free_operator,
            aggregation_type=aggregation_type,
            granularity=granularity,
            window_size=window_size,
            ignore_no_data=True,
        )

        operators[band] = monthly_s2_cloud_free_operator
    
    return operators
    


sentinel_bands = ["B02", "B03", "B04", "B08", "NDVI"]
operators = multiple_s2_cloud_free_bands(sentinel_bands, granularity="months", window_size=1, aggregation_type="mean")

projected_points = geoengine.unstable.workflow_operators.Reprojection(ge_point_source, target_spatial_reference="EPSG:32632")

points_with_s2_cloud_free = geoengine.unstable.workflow_operators.RasterVectorJoin(
    raster_sources=[x for x in operators.values() ],
    vector_source=projected_points,
    new_column_names=[x for x in operators.keys()],
)

points_with_s2_cloud_free.to_dict()
        
```

```python
workflow = geoengine.register_workflow(points_with_s2_cloud_free.to_workflow_dict())
workflow
```

```python
import datetime

start_dt = datetime.datetime(2021, 4, 1, 0, 0, 0, tzinfo=datetime.timezone.utc)
end_dt = datetime.datetime(2021, 10, 30, 0, 0, 0, tzinfo=datetime.timezone.utc)


gp_res = await workflow.vector_stream_into_geopandas(
    geoengine.QueryRectangle(
        spatial_bounds=geoengine.BoundingBox2D(
            xmin=xmin,
            ymin=ymin,
            xmax=xmax,
            ymax=ymax,
        ),
        time_interval=geoengine.TimeInterval(
            start=start_dt,
            end=end_dt,
        ),
        resolution=geoengine.SpatialResolution(
            100.0,
            100.0,
        ),
        srs="EPSG:32632",
))

gp_res.to_file("gp_res_100_frac1.gpkg", driver="GPKG")
gp_res
```
