```python
from datetime import datetime, timedelta

def generate_props_and_meta(band):

  upper_left_x_coord = 650000
  upper_left_y_coord = 9600000
  pixel_size = 10
  with_in_pixels = 8000
  height_in_pixels = 8000
  data_type = "U16"


  lower_right_x_coord = upper_left_x_coord + (pixel_size * with_in_pixels)
  lower_right_y_coord = upper_left_y_coord - (pixel_size * height_in_pixels)

  time_start_dt = datetime.strptime(start_times[0]['time'], "%Y-%m-%dT%H-%M-%S%z")
  time_end_dt = datetime.strptime(start_times[-1]['time'], "%Y-%m-%dT%H-%M-%S%z") # TODO + 14 days?
  time_end_dt = time_end_dt + timedelta(weeks=4)

  dataset_id = band['dataset_id']
  dataset_name = band['name']
  dataset_Description = band['name']
  measurement = band['measurement']
  unit = band['unit']
  folder_suffix = band['folder_suffix']
  data_type = "U16"

  unscaled_colorizer = ge.LinearGradientColorizer(
          breakpoints = [
            ge.ColorBreakpoint(0.0, (68, 1, 84, 255)),
            ge.ColorBreakpoint(20000.0 , (33, 145, 140, 255)),
            ge.ColorBreakpoint(60000.0, (253, 231, 37, 255))
          ],
          no_data_color =(0, 0, 0, 0),
          default_color = (0, 0, 0, 0)
        )

  unscaled_symbology = ge.RasterSymbology(
        colorizer = unscaled_colorizer
      )

  era5_add_data_properties = ge.DatasetProperties(
      dataset_id=dataset_id,
      name = dataset_name,
      description = dataset_Description,
      source_operator = "GdalSource",
      provenance = [ge.Provenance(
        citation = "TODO",
        license = "TODO",
        uri = "TODO"
      ), ge.Provenance(
        citation = "Geo Engine GmbH",
        license = "TODO",
        uri = "TODO"
        )],
      symbology = unscaled_symbology,
  )


  result_descriptor = ge.RasterResultDescriptor(
    data_type = data_type,
    spatial_reference = "EPSG:32717",
    spatial_resolution=ge.SpatialResolution(pixel_size, pixel_size),
    spatial_bounds=ge.SpatialPartition2D(upper_left_x_coord, lower_right_y_coord, lower_right_x_coord, upper_left_y_coord),
    time_bounds=ge.TimeInterval(time_start_dt, time_end_dt),
    measurement = ge.ContinuousMeasurement(measurement = measurement, unit = unit),
  )

  params = []
  for start_time in start_times:
    time_start = start_time['time']
    time_end = start_time['end']
    time_formatted = start_time['formated']

    start_dt = datetime.strptime(time_start, "%Y-%m-%dT%H-%M-%S%z")
    end_dt = datetime.strptime(time_end, "%Y-%m-%dT%H-%M-%S%z")
    ti = ge.TimeInterval(start=start_dt, end=end_dt )
    
    params.append(
      ge.api.GdalLoadingInfoTemporalSlice(
        time = ti.to_api_dict(),
        params = ge.api.GdalDatasetParameters({
          "filePath": f'{data_path_in_volume}/{folder_suffix}/raster_{time_formatted}.tiff',
          "rasterbandChannel": 1,
        "geoTransform": {
          "originCoordinate": {
            "x": upper_left_x_coord,
            "y": upper_left_y_coord
          },
          "xPixelSize": pixel_size,
          "yPixelSize": -1.0 * pixel_size
        },
        "width": with_in_pixels,
        "height": height_in_pixels,
        "fileNotFoundHandling": "NoData",
        })        
      )
    )

  

  era5_add_data_meta_data = ge.api.GdalMetaDataList({
      "type": "GdalMetaDataList",
      "resultDescriptor": result_descriptor.to_api_dict(),
      "params": params
    })


  return (era5_add_data_properties, era5_add_data_meta_data) 
```

```python
import json
import os


for (name, band) in bands.items():

    
    
    file_name = name

    out_dir = f'./test_data/sentinel_2_ecador/'

    if not os.path.exists(out_dir):
        os.makedirs(out_dir)

    (a,m) = generate_props_and_meta(band)

    definition = ge.api.DatasetDefinition({
        'properties': a.to_api_dict(),
        'metaData': m
    })

    with open(os.path.join(out_dir, f"{file_name}.json"), "w") as out_file:
        json.dump(definition, out_file, indent=2)
        out_file.write("\n")

    
```
